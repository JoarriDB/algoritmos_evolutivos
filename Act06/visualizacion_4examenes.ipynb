```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Cargar archivo con notas generadas para 4 exmenes
df = pd.read_csv("notas_4examenes.csv")  # Asegrate de haberlo generado

# Histograma
plt.figure(figsize=(9, 5))
sns.histplot(data=df, x="Nota", hue="Examen", bins=10, kde=True, multiple="stack", palette="Set2")
plt.title("Histograma de Notas por Examen (4 Exmenes)")
plt.xlabel("Nota")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.savefig("histograma_4examenes.png")
plt.show()

# Boxplot
plt.figure(figsize=(9, 5))
sns.boxplot(data=df, x="Examen", y="Nota", palette="Set3")
plt.title("Boxplot de Notas por Examen (4 Exmenes)")
plt.xlabel("Examen")
plt.ylabel("Nota")
plt.tight_layout()
plt.savefig("boxplot_4examenes.png")
plt.show()

# Violinplot
plt.figure(figsize=(9, 5))
sns.violinplot(data=df, x="Examen", y="Nota", palette="Set1")
plt.title("Violinplot de Notas por Examen (4 Exmenes)")
plt.xlabel("Examen")
plt.ylabel("Nota")
plt.tight_layout()
plt.savefig("violinplot_4examenes.png")
plt.show()

# Estadsticas
print("\nESTADSTICAS POR EXAMEN:")
agrupado = df.groupby("Examen")["Nota"]
promedios = agrupado.mean()
varianzas = agrupado.var()

print(promedios)
print("\nVarianzas:")
print(varianzas)

std_promedios = np.std(promedios)
diferencia_maxima = promedios.max() - promedios.min()

print(f"\nDesviacin estndar entre promedios: {std_promedios:.4f}")
print(f"Diferencia mxima entre promedios: {diferencia_maxima:.2f}")
```
