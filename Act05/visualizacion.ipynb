Importación de librerías

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Cargar y graficar evolución del fitness

```python
# Cargar archivos
hist_binaria = np.load("historial_binaria.npy")
hist_real = np.load("historial_real.npy")
hist_permutacional = np.load("historial_permutacional.npy")

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(hist_binaria, label="Binaria")
plt.plot(hist_real, label="Real")
plt.plot(hist_permutacional, label="Permutacional")
plt.title("Evolución del Fitness por Generación")
plt.xlabel("Generación")
plt.ylabel("Fitness")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Función para histograma de notas

```python
def mostrar_histograma(path_csv, titulo):
    df = pd.read_csv(path_csv)
    plt.figure(figsize=(9, 5))
    sns.histplot(data=df, x="Nota", hue="Examen", bins=10, kde=True, palette="Set2", multiple="stack")
    plt.title(f"Histograma de Notas por Examen - {titulo}")
    plt.xlabel("Nota")
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()
```

Histograma por representación

```python
mostrar_histograma("notas_binaria.csv", "Binaria")
mostrar_histograma("notas_real.csv", "Real")
mostrar_histograma("notas_permutacional.csv", "Permutacional")
```

Comparar distribuciones con Boxplot

```python
# Cargar archivos
df_bin = pd.read_csv("notas_binaria.csv")
df_bin["Representacion"] = "Binaria"

df_real = pd.read_csv("notas_real.csv")
df_real["Representacion"] = "Real"

df_perm = pd.read_csv("notas_permutacional.csv")
df_perm["Representacion"] = "Permutacional"

df_todas = pd.concat([df_bin, df_real, df_perm], ignore_index=True)

# Gráfico
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_todas, x="Representacion", y="Nota", hue="Examen", palette="Set3")
plt.title("Distribución de Notas por Representación y Examen")
plt.ylabel("Nota")
plt.xlabel("Representación Cromosómica")
plt.tight_layout()
plt.show()
```
