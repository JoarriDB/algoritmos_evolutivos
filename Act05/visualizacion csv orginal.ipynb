```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Cargar archivo CSV original
df = pd.read_csv('notas_1u.csv')

# Cambiar nombre de columna si es necesario
df.rename(columns=lambda x: x.strip(), inplace=True)

# =========================
# Histograma general de notas
# =========================
plt.figure(figsize=(8, 5))
sns.histplot(df['Nota'], bins=10, kde=True, color='skyblue')
plt.title("Histograma General de Notas")
plt.xlabel("Nota")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.savefig("histograma_notas_original.png")
plt.show()

# =========================
# Boxplot por Tipo de Examen
# =========================
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Tipo_Examen', y='Nota', palette='Set2')
plt.title("Boxplot de Notas por Tipo de Examen")
plt.xlabel("Examen")
plt.ylabel("Nota")
plt.tight_layout()
plt.savefig("boxplot_notas_original.png")
plt.show()

# =========================
# Violinplot por Tipo de Examen
# =========================
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='Tipo_Examen', y='Nota', palette='Set3')
plt.title("Violinplot de Notas por Tipo de Examen")
plt.xlabel("Examen")
plt.ylabel("Nota")
plt.tight_layout()
plt.savefig("violinplot_notas_original.png")
plt.show()

# =========================
# Estadsticas
# =========================
print("\nESTADSTICAS POR EXAMEN ORIGINAL:")
agrupado = df.groupby("Tipo_Examen")["Nota"]
promedios = agrupado.mean()
varianzas = agrupado.var()

print(promedios)
print("\nVarianzas:")
print(varianzas)

std_promedios = np.std(promedios)
diferencia_maxima = promedios.max() - promedios.min()

print(f"\nDesviacin estndar entre promedios: {std_promedios:.4f}")
print(f"Diferencia mxima entre promedios: {diferencia_maxima:.2f}")
```
